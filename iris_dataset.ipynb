Import all your packages


```python
import numpy as np
```

```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

Now to load your Iris Dataset

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
iris
```

New Command! Get all rows and first four columns as numpy data

```python
iris = iris.iloc[:,:4].to_numpy()
```

```python
iris
```

Arrange cols according to their labels

```python
sepal_length = iris[:,0]
sepal_width= iris[:,1]
petal_length = iris[:,2]
petal_width = iris[:,3]
```

Scatter Plot of Petal Length and Width

```python
plt.scatter(petal_length[:50], petal_width[:50], s=5, color='blue',label='Setosa')
plt.scatter(petal_length[50:100], petal_width[50:100], s=5, color='red',label='Versicolor')
plt.scatter(petal_length[100:150], petal_width[100:150], s=5, color='green',label='Virginica')
plt.title('Iris Scatter based off Petal Dimensions')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
```

Scatter plot all four Iris Properties, visualizing Sepal length and Width color coded accordingly by size

```python
plt.scatter( petal_length, petal_width, sepal_length*10, sepal_width, )
plt.title('Iris Scatter Diagram wit Petal Dimensions on Axes and Sepal information on the Scatter')


plt.legend()
```

Histogram and Box Plot Diagram of Petal lengths - Boxplot all four measurements

```python
plt.hist(petal_length, bins = np.arange(0,8,0.5), density=True)
plt.xlabel('Petal Length')
plt.ylabel('Density')
plt.title('Petal Length Measurements')
```

```python
plt.boxplot(
    [petal_length,petal_width,sepal_length, sepal_width],
    labels=['Petal Length', 'Petal Width', 'Sepal Length', 'Sepal Width']
)
```
